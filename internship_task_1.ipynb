```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
print(df)
```

Conclusion 1: Sepal Length vs Sepal Width

The scatter plot of sepal length versus sepal width shows a weak or no clear relationship between the two variables. The data points are widely scattered, indicating that an increase in sepal length does not significantly affect sepal width. Therefore, there is no strong correlation observed between these features.

```python

sns.scatterplot(x="sepal_length", y="sepal_width", data=df)
plt.title('Sepal Length vs Sepal Width in Iris Dataset')
plt.xlabel("sepal_length (cm)")
plt.ylabel("sepal_width(cm)")
plt.show()
```

Conclusion 2: Petal Length vs Petal Width

The scatter plot of petal length versus petal width shows a strong positive relationship. As petal length increases, petal width also increases. The data points form a clear upward trend, indicating a strong positive correlation between these two variables.

```python
sns.scatterplot(x="petal_length", y="petal_width", data=df)
plt.title('petal Length vs petal Width in Iris Dataset')
plt.xlabel("petal_length (cm)")
plt.ylabel("petal_width(cm)")
plt.show()
```

Sepal Features: Both Sepal Length and Width show an approximately symmetric (Normal) distribution. Most data points are centered around the average, with Sepal Width being the most balanced.

```python
sns.histplot(df["sepal_length"])
plt.title("distributions of sepal length")
plt.xlabel("sepal_length (cm)")
plt.ylabel("Frequency/Count")
plt.show()
```

Sepal Width: Highly Symmetric and Bell-shaped; it follows a Normal Distribution centered at 3.0cm.

```python
sns.histplot(df["sepal_width"])
plt.title("distributions of width length")
plt.xlabel("sepal_width (cm)")
plt.ylabel("Frequency/Count")
plt.show()
```

Petal Length: Bimodal distribution; shows a clear gap between a small-petal group and a large-petal group.
---



```python

sns.histplot(df["petal_length"])
plt.title("Distribution of Petal Length")
plt.xlabel("petal_length (cm)")
plt.ylabel("Frequency/Count")
plt.show()
```

Petal Width: Bimodal and Right-Skewed; clearly separates the data into two distinct clusters.



```python
sns.histplot(df["petal_width"])
plt.title("Distribution of Petal width")
plt.xlabel("petal_width (cm)")
plt.ylabel("Frequency/Count")
plt.show()
```

“In this step, a box plot is used to visualize the spread of data and detect outliers in numerical features of the dataset.”

```python
# Box plot
sns.boxplot(data=df[["sepal_length", "sepal_width", "petal_length", "petal_width"]])

plt.title("Box Plot for Detecting Outliers and Spread")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```

Conclusion:
“The box plot shows the distribution of all numerical features. Few outliers are observed in sepal_width, while other features show relatively stable distributions. Overall, the data is well distributed.”
